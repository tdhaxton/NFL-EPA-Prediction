```python
import statistics

mae_play_runs1 = [0.9619, 0.9599, 0.9620]
mae_game_runs1 = [16.48, 16.70, 16.46]

mean_play_mae1 = statistics.mean(mae_play_runs1)
std_play_mae1 = statistics.stdev(mae_play_runs1)

print("Mean/play:", mean_play_mae1)
print("Standard Deviation/play:", std_play_mae1)

mean_game_mae1 = statistics.mean(mae_game_runs1)
std_game_mae1 = statistics.stdev(mae_game_runs1)

print("Mean/game:", mean_game_mae1)
print("Standard Deviation/game:", std_game_mae1)
```
